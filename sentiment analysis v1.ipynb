```python
soup = bs(driver.page_source, 'html.parser')
rlist = soup.find_all('span',class_='section-review-text')
ulist = soup.find_all('div', class_='section-review-title')
dlist = soup.find_all('span', class_='section-review-publish-date')
ratinglist = soup.find_all('span', class_='section-review-stars').find('aria-label')
```

```python
modification = [re.findall('Rated \d+\S\d+',str(m)) for m in ratinglist]
```

```python
dataset = pd.DataFrame(columns=['User Name','Date','rating','text'])
dataset['text'] = [t.text for t in rlist]
dataset['rating'] = modification[:int(len([t.text for t in rlist]))]
dataset['User Name'] = [i.text for i in ulist]
dataset['Date'] = [d.text for d in dlist]
dataset.to_csv("C:/Users/abdelsalam/Documents/sentiment analysis aqc.csv", index=None)
dataset.head(50)
```
